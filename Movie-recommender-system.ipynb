```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L   
```

```python
convert(movies.iloc[0].genres)
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=10:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L   
```

```python
convert_cast(movies.iloc[0].cast)
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L   
```

```python
convert_crew(movies.iloc[0].crew)
```

```python
movies["crew"] = movies["crew"].apply(convert_crew)
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def lemmatization(text):
    ans = []
    tokens = nlp(text)
    for token in tokens:
        ans.append(token.lemma_)
    return ' '.join(ans)
```

```python
new_df["tags"] = new_df["tags"].apply(lemmatization)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000, stop_words="english")
```

```python
tfidf_matrix = tfidf.fit_transform(new_df["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(tfidf_matrix)
```

```python
similarity[1]
```

```python
similarity.shape
```

```python
sorted(enumerate(similarity[0]), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(enumerate(distances), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
